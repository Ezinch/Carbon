```python
import pandas as pd
import numpy as np
```

```python
gdp = pd.read_excel('gdp_growth.xlsx')
```

```python
gdp
```

```python
gdp.dropna(thresh=20, inplace=True)
```

```python
gdp
```

```python
gdp = gdp.melt(id_vars='Country Name', var_name='Year', value_name='GDP_g')
```

```python
gdp = gdp.sort_values(['Country Name', 'Year'], ignore_index=True)
```

```python
gdp
```

```python
infl = pd.read_excel('inf.xlsx')
```

```python
infl
```

```python
infl.dropna(thresh=20, inplace=True)
infl
```

```python
infl = infl.melt(id_vars='Country Name', var_name='Year', value_name='Inflation_g')
```

```python
infl = infl.sort_values(['Country Name', 'Year'], ignore_index=True)
```

```python
infl
```

```python
gdp_inf = pd.merge_ordered(gdp, infl)
```

```python
gdp_inf
```

```python
gdp_inf['True Growth'] = np.where(gdp_inf['GDP_g']>gdp_inf['Inflation_g'], 1, 0)
```

```python
gdp_inf
```

```python
gdp_inf = gdp_inf.rename(columns={'Country Name':'Country'})
```

```python
gdp_inf.to_csv('gdp.csv',index=False)
```
