```python
import pandas as pd
```

```python
alt_en = pd.read_excel('alt_energy_use.xlsx')
```

```python
alt_en
```

```python
alt_en = alt_en.melt(id_vars='Country Name', var_name='Year', value_name='Alt. en. use')
```

```python
alt_en = alt_en.sort_values(['Country Name', 'Year'], ignore_index=True)
```

```python
alt_en
```

```python
ghg = pd.read_excel('ghg.xlsx')
```

```python
ghg = ghg.melt(id_vars='Country Name', var_name='Year', value_name='GHG')
```

```python
ghg = ghg.sort_values(['Country Name', 'Year'], ignore_index=True)
```

```python
overall_en = pd.read_excel('energy_use.xlsx')
```

```python
overall_en = overall_en.melt(id_vars='Country Name', var_name='Year', value_name='Overall en. use')
```

```python
overall_en = overall_en.sort_values(['Country Name', 'Year'], ignore_index=True)
```

```python
others = pd.merge_ordered(ghg, overall_en)
```

```python
others
```

```python
others = pd.merge_ordered(others, alt_en)
```

```python
others
```

```python
others.to_csv('others.csv',index=False)
```
