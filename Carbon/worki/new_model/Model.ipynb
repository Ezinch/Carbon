```python
import pandas as pd
```

```python
rdd = pd.read_csv('rdd22.csv')
rdd
```

```python
co2 = pd.read_excel('co2_2.xlsx')
co2
```

```python
df = pd.merge_ordered(co2, rdd)
df
```

```python
df.dropna(inplace=True, ignore_index = True)
df
```

```python
df.to_excel('data.xlsx', index=False)
```

```python
year = pd.Categorical(df.Year)
```

```python
df = df.set_index(["Country", "Year"])
```

```python
exog_vars = df.columns.tolist()
exog_vars.pop(0)
exog_vars
```

```python
import statsmodels.api as sm
```

```python
exog = sm.add_constant(df[exog_vars])
```

```python
from linearmodels.panel import PooledOLS
```

```python
mod = PooledOLS(df.Emission, exog)
```

```python
pooled_res = mod.fit()
```

```python
pooled_res
```

```python
from linearmodels.panel import RandomEffects
```

```python
mod = RandomEffects(df.Emission, exog)
```

```python
re_res = mod.fit()
re_res
```

```python
from linearmodels.panel import BetweenOLS
```

```python
mod = BetweenOLS(df.Emission, exog)
be_res = mod.fit(reweight=True)
be_res
```

```python
from linearmodels.panel import compare
```

```python
print(compare({"BE": be_res, "RE": re_res, "Pooled": pooled_res}, stars=True))
```

```python
from statsmodels.stats.diagnostic import het_breuschpagan
```

```python
residuals = re_res.resids
het_test = het_breuschpagan(residuals, exog_het=exog, robust=False)
het_test
```

```python
residuals = pooled_res.resids
het_test = het_breuschpagan(residuals, exog_het=exog, robust=False)
het_test
```

```python
import numpy as np
import numpy.linalg as la
from scipy import stats
```

```python
def hausman(fe, re):
    b = fe.params
    B = re.params
    v_b = fe.cov
    v_B = re.cov
    df = b[np.abs(b) < 1e8].size
    chi2 = np.dot((b - B).T, la.inv(v_b - v_B).dot(b - B)) 
    pval = stats.chi2.sf(chi2, df)
    return chi2, df, pval
```

```python
hausman(be_res, re_res)
```
