```python
import pandas as pd
import numpy as np
```

```python
ghg = pd.read_excel("GHG.xlsx")
```

```python
ghg
```

```python
ghg.dropna(thresh=20, inplace=True)
```

```python
ghg
```

```python
ghg_nums = ghg.iloc[:, 1:]
ghg_nums.dtypes
```

```python
for i in ghg_nums.columns.tolist():
    ghg_nums[i] = ghg_nums[i].replace(',', '.', regex=True)
    ghg_nums[i] = ghg_nums[i].replace(' ', '', regex=True)
    ghg_nums[i] = ghg_nums[i].replace('-', np.NaN, regex=True)
    ghg_nums[i] = ghg_nums[i].astype('float64')
ghg_nums.dtypes
```

```python
ghg_nums.mean(numeric_only=True)
```

```python
ghg_nums.mean(axis=None, numeric_only=True)
```

```python
ghg_nums.median(numeric_only=True)
```

```python
ghg_nums.std(numeric_only=True)
```

```python
ghg_nums.min(numeric_only=True)
```

```python
ghg_nums.max(numeric_only=True)
```

```python
data_ghg_years = {'years' : ghg_nums.columns.tolist(),
                  'mean' : ghg_nums.mean(numeric_only=True).tolist(),
                  'median' : ghg_nums.median(numeric_only=True).tolist(),
                  'standard deviation' : ghg_nums.std(numeric_only=True).tolist(),
                  'minimum' : ghg_nums.min(numeric_only=True).tolist(),
                  'maximum' : ghg_nums.max(numeric_only=True).tolist()
                 }
```

```python
ghg_years = pd.DataFrame(data=data_ghg_years)
```

```python
ghg_years
```

```python
data_ghg_countries = {'country' : ghg['Страна'].tolist(),
                  'mean' : ghg_nums.mean(axis=1, numeric_only=True).tolist(),
                  'median' : ghg_nums.median(axis=1, numeric_only=True).tolist(),
                  'standard deviation' : ghg_nums.std(axis=1, numeric_only=True).tolist(),
                  'minimum' : ghg_nums.min(axis=1, numeric_only=True).tolist(),
                  'maximum' : ghg_nums.max(axis=1, numeric_only=True).tolist()
                 }
```

```python
ghg_countries = pd.DataFrame(data=data_ghg_countries)
```

```python
ghg_countries
```

```python
data_ghg_overall = {'mean' : ghg_nums.mean(axis=None, numeric_only=True),
                  'median' : ghg_nums.median(axis=None, numeric_only=True),
                  'standard deviation' : ghg_nums.std(numeric_only=True).mean(),
                  'minimum' : ghg_nums.min(axis=None, numeric_only=True),
                  'maximum' : ghg_nums.max(axis=None, numeric_only=True)
                 }
```

```python
ghg_overall = pd.DataFrame(data=data_ghg_overall, index=[0])
```

```python
ghg_overall
```

```python
with pd.ExcelWriter('Descriptive_statistics_of_raw_data.xlsx') as writer:  
    ghg_years.to_excel(writer, sheet_name='ghg_years')
    ghg_countries.to_excel(writer, sheet_name='ghg_countries')
    ghg_overall.to_excel(writer, sheet_name='ghg_overall')
```
