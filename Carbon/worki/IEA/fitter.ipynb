```python
import pandas as pd
from pathlib import Path

p = Path('')
files = []
for x in p.rglob("*"):
    if str(x).find('.xlsx') != -1:
        files.append(str(x))
files
```

```python
output = 0
for i in range(len(files)):
    file = pd.read_excel(files[i])
    file = file.melt(id_vars = 'Country', var_name = 'Year', value_name = files[i].replace('.xlsx', ''))
    file = file.sort_values(['Country', 'Year'], ignore_index=True)
    file.dropna(inplace=True, ignore_index = True)
    file = file.groupby(['Country', 'Year'])[files[i].replace('.xlsx', '')].sum().reset_index()
    if i == 0:
        output = file
    else:
        output = pd.merge_ordered(output, file)
output.dropna(inplace=True, ignore_index = True)
output
```

```python
gdp = pd.read_csv('gdp.csv')
```

```python
overall = pd.merge_ordered(gdp, output)
overall
```

```python
overall.dropna(inplace=True, ignore_index = True)
overall
```

```python
overall.drop(columns=['True Growth'], inplace = True)
```

```python
year = pd.Categorical(overall.Year)
overall = overall.set_index(["Country", "Year"])
```

```python
overall['GDP_lag'] = overall.groupby('Country')['GDP_g'].shift(1)
overall.dropna(inplace=True)
overall
```

```python
exog_vars = overall.columns.tolist()
exog_vars.pop(0)
exog_vars
```

```python
import statsmodels.api as sm
exog = sm.add_constant(overall[exog_vars])
```

```python
from linearmodels.panel import PooledOLS
```

```python
mod = PooledOLS(overall.GDP_g, exog)
pooled_res = mod.fit()
pooled_res
```

```python
from linearmodels.panel import RandomEffects
```

```python
mod = RandomEffects(overall.GDP_g, exog)
re_res = mod.fit()
re_res
```

```python
from linearmodels.panel import BetweenOLS
```

```python
exog = sm.add_constant(overall[exog_vars])
mod = BetweenOLS(overall.GDP_g, exog)
be_res = mod.fit(reweight=True) #панели несбалансированы?
be_res
```

```python
from linearmodels.panel import compare
```

```python
print(compare({"BE": be_res, "RE": re_res, "Pooled": pooled_res}, stars=True))
```

```python
import numpy.linalg as la
import numpy as np
from scipy import stats
```

```python
def hausman(fe, re):
    b = fe.params
    B = re.params
    v_b = fe.cov
    v_B = re.cov
    df = b[np.abs(b) < 1e8].size
    chi2 = np.dot((b - B).T, la.inv(v_b - v_B).dot(b - B)) 
    pval = stats.chi2.sf(chi2, df)
    return chi2, df, pval
```

```python
hausman(be_res, re_res)
```
