```python
import pandas as pd
```

```python
laws = pd.read_csv("net_zero_content.csv")
```

```python
laws
```

```python
laws.drop(columns=['ISO', 'Global category', 'Overview category', 'Sector', 'Subsector', 'Indicator ID', 'Source', 'Indicator name'], inplace=True)
laws
```

```python
import re
```

```python
for i in range(len(laws['Value'])):
    laws['Value'][i] = re.sub(r'[а-яА-Яa-zA-Z]', '', laws['Value'][i])
    laws['Value'][i] = laws['Value'][i].replace('(', '').replace(')', '').replace('<>', '').replace(',', '')
```

```python
laws = pd.read_csv('paws.csv')
laws
```

```python
import datetime
for i in range(len(laws['Value'])):
    dates = laws['Value'][i]
    date_list = dates.split()
    date_list = sorted(date_list, key=lambda x: datetime.datetime.strptime(x, '%m/%d/%Y'))
    laws['Value'][i] = date_list[0]
laws
```

```python
laws.info()
```

```python
for i in range(len(laws['Value'])):
    laws['Value'][i] = laws['Value'][i].replace('/', '')
    laws['Value'][i] = int(laws['Value'][i]) % 10000
laws
```
